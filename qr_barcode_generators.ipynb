```python
import cv2
import barcode
from barcode.writer import ImageWriter
import qrcode
from  matplotlib import pyplot as plt
from PIL import Image
```

```python
def qr_generator(data: str, name: str):
    
    qr = qrcode.QRCode(
        version=1,
        error_correction=qrcode.constants.ERROR_CORRECT_H,
        box_size=1,
        border=1,
    )
    qr.add_data(data)
    qr.make(fit=True)

    qr_img = qr.make_image(back_color=(10, 10, 70), fill_color=(255, 255, 50))
    qr_img.save(name)
```

```python
def qr_fast_generator(data: str, name: str):

    qr = qrcode.make(data)
    qr.save(name)
```

```python
def barcode_generator(data: str, type_of_coding: str, name: str):
    
    barcode_format = barcode.get_barcode_class(type_of_coding)
    my_barcode = barcode_format(data, writer=ImageWriter())
    my_barcode.save(name)
```

```python
data = 'Cremator'
path = 'qr.png'
qr_generator(data, path)

img = Image.open(path)
plt.figure(dpi=150)
plt.imshow(img)
plt.axis('off') 
plt.title(label=data, fontsize=10)
plt.show()
```

```python
data = 'Cremator'
path = 'qr_fast.png'
qr_fast_generator(data, path)

img = Image.open(path)
plt.figure(dpi=150)
plt.imshow(img)
plt.axis('off') 
plt.title(label=data, fontsize=10)
plt.show()
```

```python
data = 'Cremator'
path = 'bc'
barcode_generator(data, 'code128', path)

img = Image.open(f'{path}.png')
plt.figure()
plt.imshow(img)
plt.axis('off') 
plt.show()
```
